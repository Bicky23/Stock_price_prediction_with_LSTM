```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
```

```python
df1 = pd.read_csv('coal_daily.csv', parse_dates=[0])
df1.index = df1['Date']
df1.drop('Date', inplace=True, axis=1)
df1.head()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df1['Actual'], color='green')
plt.plot(df1['Predicted'], color='black')
plt.xticks(rotation=45)
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('Predicted vs Actual Stock Price')
plt.legend()
plt.ylim(150, 350)
plt.savefig('../coal_daily.png')
```
